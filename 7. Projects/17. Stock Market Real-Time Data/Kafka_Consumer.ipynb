```python
from kafka import KafkaConsumer
from time import sleep
from json import dumps,loads
import json
import boto3
```

```python
consumer = KafkaConsumer(
    'demo_testing2',
     bootstrap_servers=['3.235.226.170:9092'],
    value_deserializer=lambda x: loads(x.decode('utf-8')))
```

```python
#for c in consumer:
#    print(c.value)
```

```python
s3 = boto3.client("s3")
```

```python
for count, i in enumerate(consumer):
    s3.put_object(
        Bucket="kafka-real-time-data-pipeline",
        Key=f"stock_market_{count}.json",
        Body=json.dumps(i.value)
    )
```
