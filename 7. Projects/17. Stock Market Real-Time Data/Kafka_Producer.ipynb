```python
import pandas as pd
from kafka import KafkaProducer
from time import sleep
from json import dumps
import json
```

```python
producer = KafkaProducer(bootstrap_servers=['3.235.226.170:9092'], 
                         value_serializer=lambda x: 
                         dumps(x).encode('utf-8'))
```

```python
producer.send('demo_testing2', value={'affan':'affan'})
```
