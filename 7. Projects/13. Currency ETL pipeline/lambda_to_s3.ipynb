```python
import requests
import pandas as pd
import boto3
import time
import datetime
```

```python
from datetime import datetime, date, timedelta

# Get current date and time
current_datetime = datetime.now()
print(f"Current datetime: {current_datetime}")
```

```python
url = "https://rest.coinapi.io/v1/exchangerate/USD?apikey=75a21776-9880-47a6-a97b-5271c1512da4&invert=true&output_format=json"

```

```python
def currencyExchnageApi():
    response = requests.get(url)
    result = response.json()
    return result

    
    
```

```python
currencyData = {
  "Date":[],
  "Symbol" : [],
  "Rate":[]
}
```

```python
data = currencyExchnageApi()

for item in data:
    if item == "rates":
    #    print(len(data[f"{item}"]))
       for data in data[f"{item}"]:
        #    print(data)
           currencyData['Date'].append(data['time'])
           currencyData['Symbol'].append(data['asset_id_quote'])
           currencyData['Rate'].append(data['rate'])

final_exchange_data = pd.DataFrame(currencyData)
final_exchange_data.to_csv(f'{current_datetime}.csv',index=False)
```
